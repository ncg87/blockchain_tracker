```python
swap_dict = {}
for event in results:
    if event['factory_address'] not in swap_dict:
        swap_dict[event['factory_address']] = [event]
    else:
        swap_dict[event['factory_address']].append(event)

```

```python
len(swap_dict)
```

```python
swap_numbers_dict = {}
for key, value in swap_dict.items():
    swap_numbers_dict[key] = {'Ethereum': 0, 'Base': 0, 'BNB': 0}
    ethereum = 0
    base = 0
    bnb = 0
    for swap in value:
        if swap['network'] == 'Ethereum':
            swap_numbers_dict[key]['Ethereum'] += 1
        elif swap['network'] == 'Base':
            swap_numbers_dict[key]['Base'] += 1
        elif swap['network'] == 'BNB':
            swap_numbers_dict[key]['BNB'] += 1
    print("Factory:", key, "Ethereum:", swap_numbers_dict[key]['Ethereum'], "Base:", swap_numbers_dict[key]['Base'], "BNB:", swap_numbers_dict[key]['BNB'])

```
